```python
from ultralytics import YOLO

model = YOLO(model='Models/yolov8x.pt') # load the yolo model
run_type = 'video'


if run_type=='image':
    # when save=True is used, the image will be saved in the same directory run/predict/predicted_image
    result = model.predict("Media/image.png", save=True) # predict the image and save the image
    print(result)
    print("="*30)
    for box in result[0].boxes:
        print(box)

elif run_type=='video':
    result = model.predict("Media/input_video.mp4", save=True) # predict the image and save the image
    print(result)
    print("="*30)
    for box in result[0].boxes:
        print(box)

else:
    pass
```

```python
# .predict is generally for images. but for video, we need to track each object while the video progresses. if we use predict, we will need to stitch the bounding boxes of detected objects in each frame to get the tracking. that is too much work. so we use .track instead of .predict
model = YOLO(model='Models/yolov8x.pt')
result = model.track("Media/input_video.mp4", save=True) # track the objects in the video and save the video
```
